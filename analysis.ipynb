Make a method to plot some question and the scores for each of the jobs:

```python
def plot_group_score_by_question(df, question):
    # Get the question df
    df_question = df[df['Question Label'] == question]
    print(df_question['Ordforklaring'].values[0])
    print("\n")
    
    # Get the scores (without total) and sort them
    df_scores = df_question[df_question['Group'] != 'Total'][['Score', 'Group']].sort_values('Score',ascending=False)
    score_values = df_scores['Score'].values
    
    # Plot the figure
    plt.figure(figsize = (7, 10))
    plt.rcParams.update({'font.size': 8})
    plt.margins(y = 0.01)
    plt.barh(df_scores['Group'].values, score_values, height = 0.6, align = 'center')
    plt.xlabel("Score", fontsize=10)
    plt.ylabel("Brancher/jobs", fontsize=10)
    
    mean = str(round(np.mean(score_values),2))
    std = str(round(np.std(score_values),2))
    title = 'qpx --- $\mu = mpx, \sigma = spx$'.replace("qpx", question).replace("mpx", mean).replace("spx", std)   
    #plt.title(question, fontsize=14)
    plt.title(title, fontsize=16)


    
    # Return the specific scores for the question, sorted by group. Useful later for correlation
    return df_scores.sort_values(by='Group')
```

Make method to plot regplots for two sets of questions:

```python
def scatterplot_scores(df, sc1_data, sc2_data, sc1_title, sc2_title):
    # Test that the groups and order is the same in each list, otherwise its a bad plot
    assert(sc1_data.Group.values.tolist() == sc2_data.Group.values.tolist())
    
    # Format the regplot & calculate correlation:
    sns.regplot(x=sc1_data.Score.values, y=sc2_data.Score.values, line_kws={"color": "red","lw":1})
    plt.rcParams['text.usetex'] = True
    plt.xlabel(sc1_title, fontsize=12)
    plt.ylabel(sc2_title, fontsize=12)
    plt.grid(alpha=0.3)
    coeff = np.corrcoef(sc1_data.Score.values, sc2_data.Score.values)
    plt.title(r'$\rho_{X,Y} = s1 $'.replace("s1",f"{round(coeff[0][1],2)}"), fontsize=16)
    
    # Also print the pearson correlation coefficients:
    print(coeff)
```

OK, now we dive into a PCA. 

```python
df_pca = df_brancher[df_brancher['Group'] != 'Total']
```

```python
all_questions = df_pca['Question Label'].value_counts().index
values = df_pca['Question Label'].value_counts().values
```

Below we can either look at all questions when doing PCA, or simply some interesting ones. In the findings all are used to utilize all aspects of the data. 

```python
values_idx = np.where(values == values[0]) # So we take the questions which have 75 answers
questions = all_questions[values_idx].tolist()
#questions = ['Stress', 'Depressive symptomer', 'Overarbejde', 'Dit arbejde er vigtigt for dig', 
#             'Søvnproblemer', 'Træt i løbet af dagen', 'Højt tempo', 'Svære tidsfrister']
```

```python
df_pca = df_pca[df_pca['Question Label'].isin(questions)]
df_pca = df_pca[['Group', 'Question Label','Score']]
groups = df_pca['Group'].unique().tolist()
```

```python
df_pca
```

Run through each group and fetch all their scores. Afterwards we make a dataframe where each row is a group with all of its scores to the questions:

```python
group_values = []
group_labels = -1
for group in groups:
    group_df = df_pca[df_pca['Group'] == group]
    group_values.append([group] + group_df.Score.values.tolist())
    group_labels = group_df['Question Label'].values

df_formatted_branche = pd.DataFrame(np.array(group_values), columns=['Group']+group_labels.tolist())
```

```python
df_formatted_branche
```

Remove some outliers, questions that contains NaN:

```python
# These two contains NaN
df_formatted_branche = df_formatted_branche.drop('Stress både fra arbejdet og fra privatlivet blandt personer med stress',axis=1)
df_formatted_branche = df_formatted_branche.drop('Kun stress fra arbejdet blandt personer med stress',axis=1)
```

Format our X matrix (the data) properly, 0 mean and unit variance. 

```python
X = df_formatted_branche.drop('Group', axis=1).to_numpy().astype('float64')
X_mean = X.mean(axis=0)
X_std = X.std(axis=0)
X = (X-X_mean)/X_std
y = df_formatted_branche[['Group']].to_numpy()
X.shape, y.shape
```

Do the PCA:

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2).fit(X)
z = pca.transform(X) 
```

```python
pca.explained_variance_ratio_[:3]
```

So ~ 45% explained by the first 2 PC's. Lets see them visualized:

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=z[:,0], y=z[:,1], s=50)

plt.title("PCA of jobs with all scores", fontsize=14)
plt.xlabel("X @ PC1", fontsize=12)
plt.ylabel("X @ PC2", fontsize=12)
#plt.axhline(y = 0, color = 'gray', linestyle = '-', linewidth=1.4, alpha=0.4)
#plt.axvline(x = 0, color = 'gray', linestyle = '-', linewidth=1.4, alpha=0.4)

plt.grid(alpha=0.3)
```

Now we fit a Kmeans:

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=5).fit(z)
```

Plot with the predicted clusters as labels:

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=z[:,0], y=z[:,1], hue=model.labels_, legend='full',
                palette=sns.color_palette("hls", 5), s=50)
plt.xlabel("X @ PC1", fontsize=12)
plt.ylabel("X @ PC2", fontsize=12)
plt.title("PCA of jobs with all scores", fontsize=14)
#plt.axhline(y = 0, color = 'gray', linestyle = '-', linewidth=1.4, alpha=0.4)
#plt.axvline(x = 0, color = 'gray', linestyle = '-', linewidth=1.4, alpha=0.4)

plt.grid(alpha=0.3)
```

Changes in tendencies over time from 2012-2018;

```python
years = ['2012', '2014', '2016', '2018']
df_gender_depri_years = []
df_gender_soevn_years = []
```

The below code essentially loops through all of the data files and formats them nicely into a single dataframe. 

```python
float_columns = ['Score aller andel']
for year in years:
    df_depri = pd.read_csv(f"./data/koen_depri_{year}.csv", sep=",", encoding="utf-8")
    df_soevn = pd.read_csv(f"./data/koen_soevproblem_{year}.csv", sep=",", encoding="utf-8")
    
    df_depri = format_float_columns(df_depri, float_columns)
    df_soevn = format_float_columns(df_soevn, float_columns)
    
    for gender in ['Kvinder', 'Mænd']:
        depri_score = round(df_depri[df_depri['Køn'] == gender]['Score aller andel'].mean(), 2)
        depri_personer = round(df_depri[df_depri['Køn'] == gender]['Antal personer'].mean(), 2)
        depri_sum = {'Gruppe': f"{gender} avg", 'Køn': gender, 'Score aller andel': depri_score, 
                     'Emne': df_depri.iloc[0]['Emne'], 'År': year, 'Antal personer': depri_personer, 
                     'Spørgsmålsformulering': df_depri.iloc[0]['Spørgsmålsformulering']}
        df_depri = df_depri.append(depri_sum, ignore_index=True)
        
        soevn_score = round(df_soevn[df_soevn['Køn'] == gender]['Score aller andel'].mean(), 2)
        soevn_personer = round(df_soevn[df_soevn['Køn'] == gender]['Antal personer'].mean(), 2)
        soevn_sum = {'Gruppe': f"{gender} avg", 'Køn': gender, 'Score aller andel': soevn_score, 
                     'Emne': df_soevn.iloc[0]['Emne'], 'År': year, 'Antal personer': soevn_personer, 
                     'Spørgsmålsformulering': df_soevn.iloc[0]['Spørgsmålsformulering']}
        df_soevn = df_soevn.append(soevn_sum, ignore_index=True)
        

    
    df_gender_depri_years.append(df_depri)
    df_gender_soevn_years.append(df_soevn)
    
df_gender_depri_years = pd.concat(df_gender_depri_years)
df_gender_soevn_years = pd.concat(df_gender_soevn_years)
df_gender_years = pd.concat([df_gender_depri_years, df_gender_soevn_years], ignore_index=True)
```

```python
df_gender_years
```

```python
def plot_gender_over_years(df, subject, male_group, female_group):
    
    # We get the dataframe with the wanted question (subject)
    df = df[df['Emne'] == subject]
    
    # Fetch all males and females
    male = df[df['Gruppe'] == male_group][['Score aller andel','År', 'Antal personer']]
    female = df[df['Gruppe'] == female_group][['Score aller andel','År', 'Antal personer']]
    
    # We do a lineplot (to show development)
    # We also do a scatterplot to show growth in the points
    sns.lineplot(x=male['År'].values, y=male['Score aller andel'].values)
    sns.scatterplot(x=male['År'].values, y=male['Score aller andel'].values, marker='o', 
                 label='Mænd', size=male['Antal personer'].values)
    sns.lineplot(x=female['År'].values, y=female['Score aller andel'].values)
    sns.scatterplot(x=female['År'].values, y=female['Score aller andel'].values, marker='o', 
                 label='Kvinder', size=female['Antal personer'].values)
    
    # Formatting nicely
    plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
    plt.title(subject, fontsize=12)
    plt.ylabel('Score', fontsize=10)
    plt.xlabel('Years', fontsize=10)
    plt.grid(alpha=0.3)
    
```

```python
plot_gender_over_years(df_gender_years, 'Depressive symptomer', 'Mænd avg', 'Kvinder avg')
```

```python
plot_gender_over_years(df_gender_years, 'Træthed og søvnproblemer', 'Mænd avg', 'Kvinder avg')
```
